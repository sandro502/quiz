```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
# ბაზასთან დაკავშირება
conn = sqlite3.connect("cars.sqlite")
cursor = conn.cursor()
print(" ბაზასთან კავშირი დამყარებულია")
```

```python
# ცხრილის შექმნა (თუ არ არსებობს)
cursor.execute("""
CREATE TABLE IF NOT EXISTS cars (
    name TEXT,
    description TEXT,
    make TEXT,
    model TEXT,
    year INTEGER,
    price INTEGER,
    engine TEXT,
    cylinders TEXT,
    fuel TEXT,
    mileage TEXT,
    transmission TEXT,
    trim TEXT,
    body TEXT,
    doors TEXT,
    exterior_color TEXT,
    interior_color TEXT,
    drivetrain TEXT
);
""")
conn.commit()
print(" ცხრილი 'cars' შეიქმნა ან უკვე არსებობს.")
```

```python
# მონაცემების წაკითხვა კრიტერიუმით
cursor.execute("SELECT name, price FROM cars WHERE price > 70000")
expensive_cars = cursor.fetchall()
print("\n 70,000-ზე მეტი ღირებულების მანქანები:")
for car in expensive_cars:
    print(car)
```

```python
# მონაცემის დამატება მომხმარებლისგან
name = input("შეიყვანე მანქანის სრული სახელი: ")
description = input("აღწერა: ")
make = input("მწარმოებელი: ")
model = input("მოდელი: ")
year = int(input("წელი: "))
price = int(input("ფასი: "))
engine = input("ძრავი: ")
cylinders = input("ცილინდრები: ")
fuel = input("საწვავი: ")
mileage = input("გარბენი: ")
transmission = input("გადაცემათა კოლოფი: ")
trim = input("კომპლექტაცია: ")
body = input("კორპუსი: ")
doors = input("კარები: ")
exterior_color = input("გარე ფერი: ")
interior_color = input("შიდა ფერი: ")
drivetrain = input("ძრავის ამძრავი სისტემა: ")
cursor.execute('''
    INSERT INTO cars (name, description, make, model, year, price, engine, cylinders, fuel, mileage, transmission, trim, body, doors, exterior_color, interior_color, drivetrain)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)''',
    (name, description, make, model, year, price, engine, cylinders, fuel, mileage, transmission, trim, body, doors, exterior_color, interior_color, drivetrain))

conn.commit()
print(" მონაცემი წარმატებით დაემატა")
```

```python
# მონაცემის განახლება rowid-ით
print("\n ამჟამინდელი ჩანაწერები:")
df_with_id = pd.read_sql_query("SELECT rowid, * FROM cars", conn)
print(df_with_id)

car_id = int(input("შეიყვანე მანქანის ID (rowid), რომლის ფასის განახლებაც გსურს: "))
new_price = int(input("შეიყვანე ახალი ფასი: "))
cursor.execute("UPDATE cars SET price = ? WHERE rowid = ?", (new_price, car_id))
conn.commit()
print(" ფასი განახლდა წარმატებით")
```

```python
# მონაცემის წაშლა მოდელით
model_to_delete = input("შეიყვანე მოდელი, რომლის წაშლა გსურს: ")
cursor.execute("DELETE FROM cars WHERE model = ?", (model_to_delete,))
conn.commit()
print(" მონაცემი წაიშალა წარმატებით")
```

```python
# განახლებული მონაცემების წაკითხვა
df = pd.read_sql_query("SELECT * FROM cars", conn)
```

```python
# mileage-ის ტიპის გარდაქმნა (სტრინგიდან რიცხვზე)
df["mileage"] = pd.to_numeric(df["mileage"], errors='coerce')
```

```python
# საშუალო ფასი მწარმოებლის მიხედვით
plt.figure(figsize=(10,6))
df.groupby("make")["price"].mean().sort_values().plot(kind='bar', color='skyblue')
plt.title("საშუალო ფასი მწარმოებლის მიხედვით")
plt.xlabel("მწარმოებელი")
plt.ylabel("საშუალო ფასი")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# გარბენის განაწილება
plt.figure(figsize=(8,5))
plt.hist(df["mileage"].dropna(), bins=15, color='orange', edgecolor='black')
plt.title("გარბენის განაწილება")
plt.xlabel("გარბენი (მილი)")
plt.ylabel("მანქანების რაოდენობა")
plt.grid(True)
plt.show()
```

```python
 # ცილინდრების რაოდენობა
plt.figure(figsize=(6,4))
df["cylinders"].value_counts().sort_index().plot(kind='bar', color='purple')
plt.title("ცილინდრების რაოდენობის განაწილება")
plt.xlabel("ცილინდრები")
plt.ylabel("მანქანების რაოდენობა")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
conn.close()
print("ბაზასთან კავშირი დაიხურა")
```
